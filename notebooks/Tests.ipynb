```python
import networkx as nx
from ipysigma import Sigma
```

```python
g = nx.karate_club_graph()
from random import random
categ = {}
for k, a in g.nodes(data=True):
    a['size'] = 20 * random()
    a['random'] = random()
    categ[k] = random() > 0.5
for _, _, a in g.edges(data=True):
    a['random'] = random()
    a['partition'] = random() > 0.5
    a['label'] = 'Test'
    
    if random() > 0.5:
        a['half'] = True

sigma = Sigma(
    g,
    node_size=g.degree,
    node_size_range=[5, 10],
    edge_label='label',
    edge_size='random',
    clickable_edges=True,
    edge_weight='random',
    edge_color_palette={True: 'yellow'},
    default_edge_color='blue',
    edge_color='half',
    node_color='club',
    node_color_palette={'Mr. Hi': 'green', 'Officer': 'pink'},
    default_edge_type='line',
    edge_sort_key=lambda s, t, a: a.get('half', 0)
)
sigma
```

```python
g = nx.read_gexf('./data/rio.gexf')

for s, t, a in g.edges.data():
    a['lang'] = g.nodes[s]['Language']

sigma = Sigma(
    g,
    node_color='Language',
    edge_size_range=5,
    clickable_edges=True,
    default_edge_type='triangle'
)
sigma
```
